```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [-1.,0.,2.]
y = [4.,1.,-1]
x1 = np.linspace(-2,4,20)
f = lambda x: 1 - 7/3*x+2/3*x**2
```

```python
plt.plot(x,y,'*r',label='dados')
plt.plot(x1,f(x1),label = '$1 - \dfrac{7}{3}x^2+\dfrac{2}{3}x^3$')
plt.title("Interpolação quadrática")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig('pol.png')
```
