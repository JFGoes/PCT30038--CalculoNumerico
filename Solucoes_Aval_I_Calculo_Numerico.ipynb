# 1 - Seja o sistema linear
$
\left\{
\begin{aligned}
    10x_1 + x_2 - x_3 &= 10\\
    2x_1 + 10 x_2 + 8x_3   &= 20\\
    7x_1 + x_2 + 10x_3  &= 30
\end{aligned}
\right.
$

#a) É possível dizer se o Método de Jacobi é convergente para esse sistema, usando o critério das linhas?

R: Sim, o método de jacobi é convergente uma vez que o critério de linha garante que o valor absoluto do pivô em cada linha é maior ou igual a soma das outras variáveis nas linhas.

```python
A = np.array([[10.0,1.,-1.], [2.,10.,8.],[7.,1.,10.]])
b = np.array([10.,20.,30.])
x = np.array([1.,1.,1.])
sol, lerro,iter = Jacobi(A,b,x,0.01,100)
```

# b) E possível dizer se o Método de Gauss-Seidel é convergente para esse sistema, usando o critério de Sassenfeld?

Sim, o método de Gauss-Seidel é convergente, uma vez que o critério de Sassenfeld é satisfeito.

```python
A = np.array([[10.0,1.,-1.], [2.,10.,8.],[7.,1.,10.]])
b = np.array([10.,20.,30.])
x = np.array([1.,1.,1.])
sol, lerro,iter = GSeidel(A,b,x,100,0.01,1)
```

# c) Resolver o sistema utilizando o Método de Gauss-Seidel com $x_0 = (0.7, −1.6, 0.6)$ e $ϵ = 10^{-2}$

# 2 - Seja o sistema linear

$
\left\{
\begin{aligned}
    x_1 + 3x_2 - x_3 &= -2\\

# 3 - Determine um intervalo (a, b), de tal forma que $∀x_0 ∈ (a, b)$ a função de iteração gere uma sequência convergente para a(s) raiz(es) de cada uma das funções abaixo, usando os método iterativo Bissecção, Newton e Secante com tolerância $ϵ = 10^{-6} $.

# a) $f(x) = \sqrt{x} - e^{-x}$

```python
y = Symbol('y')
```

```python
#Newton-Raphson
x0=0.5
erro = 1e-6
sol, lerro = NewtonRaphson(f,df,x0,erro)
```
